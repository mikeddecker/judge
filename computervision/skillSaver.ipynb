```python
import os
import psutil

DIM = 224
timesteps = 16


def process_memory():
    "Returns megabytes"
    process = psutil.Process(os.getpid())
    mem_info = process.memory_info()
    return mem_info.rss / 1024 / 1024 / 1024

print(f"Current memory usage: {process_memory()} gigabytes")
```

```python
repo = DataRepository()
frameloader = FrameLoader(repo)
repo.VideoNames.head()
```

```python
%%time

train_gen = DataGeneratorSkills(
    frameloader=frameloader,
    train_test_val="train",
    dim=(DIM,DIM),
    batch_size=1,
    timesteps=timesteps,
)
```

```python
train_gen.Skills
```

```python
train_gen.BalancedSkills3
```

```python
CROPPED_SKILLS_FOLDER = 'cropped-skills'
STORAGE_DIR = os.getenv("STORAGE_DIR")
```

```python
os.path.join(STORAGE_DIR, CROPPED_SKILLS_FOLDER)
```

```python
for i in range(len(train_gen.BalancedSkills3)):
    X, y = train_gen.__getitem__(i, normalize=False)
    row = train_gen.BalancedSkills3.iloc[i]


    id = row['id']
    videoId = row['videoId']
    frameStart = row['frameStart']
    frameEnd = row['frameEnd']
    skillpath = f"{videoId}-{frameStart}-{frameEnd}-{DIM}x{DIM}-t{timesteps}-{id}"
    skillpath

    full_path = os.path.join(STORAGE_DIR, CROPPED_SKILLS_FOLDER, f"{skillpath}.mp4")
    if os.path.exists(full_path):
        continue

    fps = timesteps / 2
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec for MP4
    out = cv2.VideoWriter(full_path, fourcc, fps, (DIM, DIM))

    for i in range(timesteps):
        out.write(X[0, i])

    out.release()
```
